```python
import pandas as pd
import os

cwd = os.getcwd()
projectDir = os.path.dirname(cwd)

#Absolute path to training data.
trainingPath = os.path.join(projectDir, "data", "raw", "train.csv")
print(trainingPath)

#Convert training data to data frame.
df = pd.read_csv(trainingPath)
X = df.iloc[:, :-1].values   #Select all rows and select all columns (except the last one). Creates NumPy matrix.
y = df.iloc[:, -1].values    #Select the last column (labels). Creates NumPy vector.

print("X shape:", X.shape)
print("y shape:", y.shape)

print("First 5 rows of X:\n", X[:5])
print("First 5 labels:\n", y[:5])
```
